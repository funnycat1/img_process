```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
img_name = '../pic/LenaRGB.tif'
```

```python
img = plt.imread(img_name)
```

```python
# plt.xticks([]) #  x 
# plt.yticks([]) #  y 
plt.figure(dpi=1200)
plt.axis('off') # 
plt.imshow(img)
```

```python
# RGB to gray 
def rgb_to_gray(img):
    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]
#     gray = (r*299 + g*587 + b*114) / 1000
#     gray = gray.astype(np.uint8)
#     
    
    gray = 0.2990*r + 0.5870*g + 0.1140*b
    
    return gray
```

```python
lena_gray = rgb_to_gray(img)
lena_gray.shape
```

```python
plt.figure(dpi=1200)
plt.axis('off') # 
plt.imshow(lena_gray, cmap='gray')
```

```python
# 
gauss_noise_img = skimage.util.random_noise(img, mode='gaussian')

plt.figure(dpi=1200)
plt.axis('off') # 
plt.imshow(gauss_noise_img)
```
