```python
import matplotlib.pyplot as plt
import numpy as np
import skimage
```

```python
img_name = '../pic/LenaRGB.tif'
```

```python
img = plt.imread(img_name)
```

```python
# plt.xticks([]) # 去掉 x 坐标，存在黑色边框
# plt.yticks([]) # 去掉 y 坐标，存在黑色边框
plt.figure(dpi=1200)
plt.axis('off') # 去掉坐标，且不存在黑色边框
plt.imshow(img)
```

```python
# RGB to gray 算法
def rgb_to_gray(img):
    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]
#     gray = (r*299 + g*587 + b*114) / 1000
#     gray = gray.astype(np.uint8)
#     上面计算的结果会整数溢出了
    
    gray = 0.2990*r + 0.5870*g + 0.1140*b
    
    return gray
```

```python
lena_gray = rgb_to_gray(img)
lena_gray.shape
```

```python
plt.figure(dpi=1200)
plt.axis('off') # 去掉坐标，且不存在黑色边框
plt.imshow(lena_gray, cmap='gray')
```

```python
# 添加高斯造影
gauss_noise_img = skimage.util.random_noise(img, mode='gaussian')

plt.figure(dpi=1200)
plt.axis('off') # 去掉坐标，且不存在黑色边框
plt.imshow(gauss_noise_img)
```
